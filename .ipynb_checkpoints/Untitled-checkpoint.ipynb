```julia
using Plots, TaylorSeries, LinearAlgebra
```

```julia
function Newton(f::Function,x0::Float64,p0::Float64,orden::Int64; ite1 = 30, tol1 = 1e-16)
    x_new = x0
    p_new = p0
    i, j, k = 1, 1, 1
    t = Taylor1(orden)
    while k <= ite1 && abs(f(x_new,p_new)) > tol1
        while i <= ite1 && abs(f(x_new,p_new)) > tol1
            x_old = x_new
            x_new = x_old - (f(x_old+t,p_new)/differentiate(f(x_old+t,p_new)))(0.0)
            i+=1
        end
        while j <= ite1 && abs(f(x_new,p_new)) > tol1
            p_old = p_new
            p_new = p_old - (f(x_new,p_old+t)/differentiate(f(x_new,p_old+t)))(0.0)
            j+=1
        end
        k += 1
    end
    #println("Las iteraciones alcanzaron: i = $(i), j = $(j), k = $(k)")
    return x_new, p_new
end
```

```julia
function Estabilidad(f::Function,x0::Float64,p0::Float64,orden::Int64,p_min::Float64,Δp::Float64,p_max::Float64,Δx::Float64;Nmax = 1000, ite = 20, tol = 1e-16)
    #N = floor(Int64,(p_max - p_min)/Δp)
    P = Float64[]
    X = Float64[]

    push!(P,p0)
    push!(X,x0)

    println("i = 0\t\t\t x = $(x0)\t p = $(p0)\t f(x,p) = $(f(x0,p0))")

    p_old = p0 + Δp
    x_old = x0

    signo = 1
    i = 1
    
    while p_min <= p_old <= p_max && i <= Nmax
        #x_test, p_test = x_old, p_old
        x_new, p_new = Newton(f,x_old,p_old,orden; ite1 = ite, tol1 = tol)
        if abs(f(x_new,p_new)) <= tol && abs(x_new - x_old) <= Δx
            #println("i = $(i),\t p_new = $(p_new)\t p_old = $(p_old)\t f(x,p) = $(f(x_new,p_new))")
            push!(P,p_new)
            push!(X,x_new)
        end
        norma = norm([P[end],X[end]] - [P[end-1],X[end-1]])
        x_old += ((X[end] - X[end-1]) / norma) * Δp
        p_old += ((P[end] - P[end-1]) / norma) * Δp
        println("i = $(i),\t p_new = $(p_new)\t p_old = $(p_old)")
        println("\t x_new = $(x_new)\t x_old = $(x_old)\t f(x,p) = $(f(x_new,p_new))")
        i+=1
    end

    return X, P
end
```

```julia
g(u,λ) = (u^2 -1)*(u^2-4) + λ*u^2*exp(u/10)
```

```julia
u0,λ0 = Newton(g,2.0,0.05,10; ite1 = 20, tol1 = 1e-16)
```

```julia
g(u0,λ0)
```

```julia
x0 = 2.0
p0 = 0.0
orden = 20
p_min = 0.0
p_max = 1.0
Δp = 0.01
Δx = 0.05

U, Λ = Estabilidad(g, x0, p0, orden, p_min, Δp, p_max,Δx)
```

```julia
plot(Λ,U)
xlims!(p_min,p_max)
ylims!(-2.5,2.5)
```
